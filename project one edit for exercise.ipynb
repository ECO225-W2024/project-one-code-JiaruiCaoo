```python
df1 = df1.dropna(subset=['Economic Need Index'])
```

```python
variable = df1['Economic Need Index']

plt.hist(variable, bins=30, alpha=0.7)
plt.title('Distribution of Economic Need Index')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
df1 = df1.dropna(subset=['Student Attendance Rate'])
```

```python
variable = df1['Student Attendance Rate']

plt.hist(variable, bins=30, alpha=0.9)
plt.title('Distribution of School Attendence Rate')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
df1['Economic Need Index'] = pd.to_numeric(df1['Economic Need Index'], errors='coerce')
df1['Student Attendance Rate'] = pd.to_numeric(df1['Student Attendance Rate'], errors='coerce')

df1 = df1.dropna(subset=['Economic Need Index', 'Student Attendance Rate'])

correlation = df1['Economic Need Index'].corr(df1['Student Attendance Rate'])
print("Correlation Coefficient:", correlation)
```
